## **Задача 5:**
### Ответьте на несколько вопросов:
### 5.1 Выберите дом клиенту. Клиент хочет дом с видом на набережную, как минимум с тремя ванными и с подвалом. Сколько вариантов есть у клиента?

```python
house_data.sample(7)
```

```python
client_choice1 = house_data[(house_data['waterfront'] != 0) & (house_data['bathrooms'] > 3) & (house_data['sqft_basement'] != 0)]
client_choice1.shape[0]
```

### 5.2 Выберите дом клиенту. Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. В каком ценовом диапазоне будут дома?

```python
client_choice2 = house_data[((house_data['view'] == 4) | (house_data['waterfront'] == 1))
                            & (house_data['condition'] == 5) & (house_data['yr_built'] >= 1980)]
client_choice2['price'].min(), client_choice2['price'].max()
```

### 5.3 Выберите дом клиенту. Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?

```python
client_choice3 = house_data[(house_data['sqft_basement'] == 0) & (house_data['floors'] == 2) & (house_data['price'] < 150000)]
client_choice3['condition'].mean()
```
